# <span style="color:blue">Compare the lists of differentially expressed genes among all the clusters by means of a </span><span style="color:red">Super Venn diagram</span><span style="color:blue">.</span>

```python
# Combine the lists of differentiallay expressed genes.
#
de_lists = [marker_genes_c0_vs_rest, marker_genes_c1_vs_rest, marker_genes_c2_vs_rest, marker_genes_c3_vs_rest,
            marker_genes_c4_vs_rest, marker_genes_c5_vs_rest, marker_genes_c6_vs_rest, marker_genes_c7_vs_rest,
            marker_genes_c8_vs_rest, marker_genes_c9_vs_rest, marker_genes_c10_vs_rest, marker_genes_c11_vs_rest,
            marker_genes_c12_vs_rest, marker_genes_c13_vs_rest, marker_genes_c14_vs_rest, marker_genes_c15_vs_rest,
            marker_genes_c16_vs_rest, marker_genes_c17_vs_rest]

# Convert the list of lists into a set.
#
de_sets = [set(element) for element in de_lists]

# Create a list with the names of the groups to be used as labels for the sets.
#
labels = ['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4', 'Cluster 5', 'Cluster 6', 'Cluster 7', 'Cluster 8',
          'Cluster 9', 'Cluster 10', 'Cluster 11', 'Cluster 12', 'Cluster 13', 'Cluster 14', 'Cluster 15', 'Cluster 16',
          'Cluster 17',]

# Generate the Venn Diagram with side plots to display the totals.
#
with plt.style.context('bmh'):
    supervenn(de_sets, labels, side_plots=True)
    plt.title("Supervenn Diagram")
    plt.xlabel("Amount of common genes between the clusters")
    plt.ylabel("Compared clusters")
    # Save the plot to an image file
    print('\nThe Super Venn diagram was saved as \"super_venn_diagram_01.png\"\n')
    plt.savefig('super_venn_diagram_01.png')
    plt.show()
    
# Generate the Venn Diagram without side plots.
#
with plt.style.context('bmh'):
    supervenn(de_sets, labels, side_plots=False)
    plt.title("Supervenn Diagram")
    plt.xlabel("Amount of common genes between the clusters")
    plt.ylabel("Compared clusters")
    # Save the plot to an image file
    print('\nThe Super Venn diagram was saved as \"super_venn_diagram_02.png\"\n')
    plt.savefig('super_venn_diagram_02.png')
    plt.show()
```

## <span style="color:blue">Bigger Super Venn diagram with side panels.</span>

```python
# Generate a bigger Super Venn diagram with side plots to display the totals.
#
plt.figure(figsize=(30, 10))
supervenn(de_sets, labels)
plt.xlabel("Amount of common genes between the clusters")
plt.ylabel("Compared clusters")
# Save the plot to an image file
print('\nThe Super Venn diagram was saved as \"super_venn_diagram_03.png\"\n')
plt.savefig('super_venn_diagram_03.png')
plt.show()
```

## <span style="color:blue">Bigger Super Venn diagram without side panels.</span>

```python
# Generate a bigger Super Venn Diagram without side plots.
#
plt.figure(figsize=(30, 10))
supervenn(de_sets, labels, side_plots=False)
plt.xlabel("Amount of common genes between the clusters")
plt.ylabel("Compared clusters")
# Save the plot to an image file
print('\nThe Super Venn diagram was saved as \"super_venn_diagram_04.png\"\n')
plt.savefig('super_venn_diagram_04.png')
plt.show()
```

# <span style="color:blue">Compare the list of differentially expressed genes from </span><span style="color:red">cluster 0 </span><span style="color:blue">vs </span><span style="color:red">cluster 2 </span><span style="color:blue">by means of a </span><span style="color:red">Venn diagram</span><span style="color:blue">.</span>

```python
print('Comparing the list of differentially expressed genes from cluster 0 vs cluster 2')

# Display marker_genes_c0_vs_rest
print('\nmarker_genes_c0_vs_rest\n')
print(marker_genes_c0_vs_rest)

# Display marker_genes_c2_vs_rest
print('\nmarker_genes_c2_vs_rest\n')
print(marker_genes_c2_vs_rest)

# Generate a Venn diagram for genes_de_g0 & genes_de_g1
venn2([set(marker_genes_c0_vs_rest), set(marker_genes_c2_vs_rest)], alpha = 1, set_labels = ('Cluster 0', 'Cluster 2'),
     set_colors=['cyan', 'red'])

# Change Backgroud
plt.gca().set_facecolor('white')
plt.gca().set_axis_on()

# Save the plot to an image file
print('\nThe Venn diagram was saved as \"venn_diagram_for_c0_vs_c2.png\"\n')
plt.savefig('venn_diagram_for_c0_vs_c2.png')

# Display the plot.
plt.show()

# Display the unique genes in marker_genes_c0_vs_rest when compared with marker_genes_c2_vs_rest
unique_c0 = list(set(marker_genes_c0_vs_rest).difference(marker_genes_c2_vs_rest))
unique_c0.sort()

print('\nUnique genes in marker_genes_c0_vs_rest when compared with marker_genes_c2_vs_rest = %d' % (len(unique_c0)))
# Display
for gene in unique_c0:
    print('\t%s' % (gene))

# Obtain the common differentially expressed genes.
common = set(marker_genes_c0_vs_rest) & set(marker_genes_c2_vs_rest)
print('\nCommon genes = %d' % (len(common)))
# Traverse common set.
for gene in common:
    print('\t', gene)
    
# Display the unique genes in marker_genes_c2_vs_rest when compared with marker_genes_c0_vs_rest
unique_c2 = list(set(marker_genes_c2_vs_rest).difference(marker_genes_c0_vs_rest))
unique_c2.sort()

print('\nUnique genes in marker_genes_c2_vs_rest when compared with marker_genes_c0_vs_rest = %d' % (len(unique_c2)))
# Display
for gene in unique_c2:
    print('\t%s' % (gene))
```
